```python
df.columns = df.columns.str.lower()
```

```python
df.columns = df.columns.str.replace(' ', '_')
```

```python
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
# purcahse columns purchase_frequency_days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 98
}
```

```python

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
from sqlalchemy import create_engine

# Step 1: PostgreSQL connection details
username = "postgres"        # default user
password = "284528"          # your PostgreSQL password
host = "localhost"           # local database
port = "5432"                # default PostgreSQL port
database = "customer_behavior"  # database name created in pgAdmin

# Create SQLAlchemy engine
engine = create_engine(
    f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}"
)

# Step 2: Load DataFrame into PostgreSQL
table_name = "customer"   # table name in PostgreSQL

df.to_sql(table_name, engine, if_exists="replace", index=False)

print(f"Data successfully loaded into table '{table_name}' in database '{database}'")
```
